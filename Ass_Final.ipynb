```python
#Question 1
#import datetime: we need to import datetime to get the current date from our local machine
# year : datetime.date.today().year will provide us the current date in YYYY format  
# obtained_year : will provide us when the individual obtained the license by subtracting
                    #the age at which the person gets his license from the date of birth
#driving_history : tells how long the driver holds the license


import datetime

print("Hollo user!")
user = input("What is your name? ")
print("Hello," + user)
x = 30
year = datetime.date.today().year

dob = int(input("What is your year of birth?: "))
#the underneath while loop controls the invalid entry
while True:
    try:
        dob = int(input("What is your year of birth?: "))
    except ValueError:
        print("Sorry, this is not year of birth.")
        continue
    if dob < 0:
        print("Sorry, your response must not be negative.")
    elif dob >= year:
        print("Invaild Entry, your response must not be greater than the current year")
        continue
    else:
        break                           #age was successfully parsed, and we're ready to exit the loop.

age = year - dob
response = input("Do you know how to drive? Please say 'yes' or 'no' ")

if  response =='yes':
    driving_license_at_age = int(input("How old were you when you got your driving license? "))
    obtained_year = dob + driving_license_at_age
    driving_history = int(year) - obtained_year
    print(f"You got your driving license in {obtained_year} and you have been driving for {driving_history} years")

if response == 'no':
    if age>30:
        print(f"You are {age} years old now but you still can't drive. You were supposed to get your license in {dob + x},{age - x} years ago at the age of {x} at the latest")
    else:
        print(f"You are {age} years old now and you should work on getting your driving license before you turn {x}. You have {x -age} years to go until the year {dob + x}")
 
```
